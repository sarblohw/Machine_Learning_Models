```python
import pandas as pd
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df.iloc[:, :1].values
y = df.iloc[:, -1:].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train, y_train, color = 'red', alpha = 0.5, label = 'Test(y)')
plt.plot(X_train, regressor.predict(X_train), color = 'black', linewidth = 2, marker = 'x', label = 'Predictions')
plt.legend(loc = 'best')
plt.xlabel('Experience (years)')
plt.ylabel('Salary (USD)')
plt.title('Prediction Line vs Testers')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'red', alpha = 0.5, label = 'Test(y)')
plt.plot(X_test, y_pred, color = 'black', linewidth = 2, marker = 'x', label = 'Predictions')
plt.legend(loc = 'best')
plt.xlabel('Experience (years)')
plt.ylabel('Salary (USD)')
plt.title('Prediction Line vs Testers')
plt.show()
```
